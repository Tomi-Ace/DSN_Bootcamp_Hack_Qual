```python
import numpy as np
import pandas as pd
pd.set_option('display.max_column', 60)
import seaborn as sbn
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
loan_train = pd.read_csv("Train.csv")
loan_test = pd.read_csv("Test.csv")
```

```python
print(f"The shape of the train dataset is {loan_train.shape} and the test data shape is given as {loan_test.shape}")
```

```python
loan_train.head()
```

```python
loan_test.head()
```

```python
loan = pd.concat((loan_train, loan_test)).reset_index(drop = True)
loan.tail()
```

```python
loan.info()
```

```python
loan.isnull().sum()
```

```python
for col in loan.drop(['Applicant_ID', 'form_field47'], axis = 1):
    loan[col].fillna(-9999, inplace=True)
```

```python
loan.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
loan['form_field47'] = label.fit_transform(loan.form_field47)
```

```python
loan.drop(['Applicant_ID', 'default_status'], axis = 1, inplace = True)
```

```python
Xtrain = loan[:56000].values
Xtest = loan[56000:].values
ytrain = loan_train.default_status.values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Xtrain, ytrain, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler, QuantileTransformer
from sklearn.ensemble import ExtraTreesClassifier, GradientBoostingClassifier
from sklearn.metrics import roc_auc_score
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
```

```python
%%time

Input = [('scale', QuantileTransformer()),('model', CatBoostClassifier(learning_rate= 0.01, n_estimators=2500, max_depth = 8))]
model = Pipeline(Input)
model.fit(Xtrain, ytrain)

pred = model.predict_proba(X_test)[:,1]
```

```python
print(f'Training score: {model.score(X_train, y_train)}')
print(f'Test score: {model.score(X_test, y_test)}')
```

```python
roc_auc_score(y_test, pred)
```

```python
%%time
model_cat = CatBoostClassifier(learning_rate = 0.01, n_estimators=2000, max_depth = 8)
model_cat.fit(X_train, y_train)

pred_cat = model_cat.predict_proba(X_test)[:,1]
```

```python
print(f'Training score: {model_cat.score(X_train, y_train)}')
print(f'Test score: {model_cat.score(X_test, y_test)}')
```

```python
roc_auc_score(y_test, pred_cat)
```

```python
%%time 

Input = [('scale', QuantileTransformer()),  ('model', LGBMClassifier(learning_rate = 0.01, n_estimators = 2500, num_leaves = 35,
        subsample = 1, subsample_freq=1, colsample_bytree = 0.45, reg_alpha = 1))]

model_light = Pipeline(Input)
model_light.fit(Xtrain, ytrain)

pred_light = model_light.predict_proba(X_test)[:,1]
```

```python
print(f'Training score: {model_light.score(X_train, y_train)}')
print(f'Test score: {model_light.score(X_test, y_test)}')
```

```python
roc_auc_score(y_test, pred_light)
```

```python
pred1 = model_cat.predict_proba(Xtest)[:,1]
pred2 = model.predict_proba(Xtest)[:,1]
pred3 = model_light.predict_proba(Xtest)[:,1]

final_pred = pred1 * 0.35 + pred2 * 0.35 + pred3 * 0.3
```

```python
submission = pd.DataFrame()
submission['Applicant_ID'] = loan_test.Applicant_ID
submission['default_status'] = final_pred
submission.head()
```

```python
submission.to_csv('submission.csv', index= False)
```
